```python
# datetime to query past year
import datetime as dt
```

```python
# pandas to convert queries to dataframes
import pandas as pd
```

```python
# matplotlib for plotting
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.orm import Session
from sqlalchemy import func, inspect
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
begin_date = dt.datetime(2017, 5, 1)
end_date = dt.datetime(2017, 5, 15)
```

```python
# link an engine to a database file called `hawaii.sqlite`
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurements
Station = Base.classes.stations
```

```python
session = Session(engine)
```

# Precipitation Analysis

```python
# Query for precipitation for every day of past year

sel = [Measurement.prcp,
       Measurement.date]

last_yr_prcp = session.query(*sel).\
    filter(Measurement.date > end_date - dt.timedelta(days=365)).\
    filter((Measurement.date) < end_date).all()
    
last_yr_prcp
```

```python
# convert query to data frame, plot data frame
prcp_df = pd.DataFrame(last_yr_prcp, columns = ['prcp', 'date'])
prcp_df.set_index('date', inplace=True)

prcp_df.plot()
plt.tight_layout()
plt.show()
```

```python
# summary of past year precipitation data
prcp_df.describe()
```

# Station Analysis

```python
sel = [func.count(Station.name)]
 
stations_count = session.query(*sel).all()[0][0]

stations_count
```

```python
# Design a query to find the most active stations.
# List the stations and observation counts in descending order
# Which station has the highest number of observations?

sel = [Measurement.station,
       func.count(Measurement.date)]

station_observations = session.query(*sel).\
                       group_by(Measurement.station).\
                       order_by(func.count(Measurement.date).desc()).all()
    
# station with most observations
highest_observation_station = station_observations[0]

highest_observation_station
```

```python
# Design a query to retrieve the last 12 months of temperature observation data (tobs).
# Filter by the station with the highest number of observations.

sel = [Measurement.tobs,
      Measurement.station,
      Measurement.date]

station_tobs = session.query(*sel).\
    filter(Measurement.date > begin_date - dt.timedelta(days=365)).\
    filter(Measurement.date < end_date).\
    filter(Measurement.station == highest_observation_station[0]).all()

station_tobs
```

```python
# Plot the results as a histogram with bins=12

top_station_df = pd.DataFrame(station_tobs, columns = ['tobs','station', 'date'])
top_station_df.set_index('date', inplace=True)
top_station_df.plot.hist(bins=12)
plt.show()
```

# Temperature Analysis

```python
# Write a function called calc_temps that will accept a start date and end date in the format %Y-%m-%d 
# and return the minimum, average, and maximum temperatures for that range of dates

def calc_temps(begin_date, end_date):
    temp_start_date = end_date - dt.timedelta(days=365)
    temp_end_date = end_date
    
    sel = [func.avg(Measurement.tobs),
          func.max(Measurement.tobs),
          func.min(Measurement.tobs)]

    temps = session.query(*sel).\
        filter(Measurement.date > temp_start_date).\
        filter(Measurement.date < temp_end_date).all()
    
    print('Temperature stats from ' + str(dt.datetime.strftime(temp_start_date, "%Y-%m-%d")) + ' to '\
          + str(dt.datetime.strftime(temp_end_date, "%Y-%m-%d")) + ':'
        + ' \nAverage: ' + str(round(temps[0][0])) + '\nMaximum: ' + str(temps[0][1]) + '\nMinimum: ' + str(temps[0][2]))
```

```python
# pass original begin and end dates through function
begin_date = dt.datetime(2017, 5, 1)
end_date = dt.datetime(2017, 5, 15)
calc_temps(begin_date, end_date)
```

```python
# repeat past year temperature query outside function

temp_start_date = end_date - dt.timedelta(days=365)
temp_end_date = end_date

sel = [func.avg(Measurement.tobs),
      func.max(Measurement.tobs),
      func.min(Measurement.tobs)]

temps = session.query(*sel).\
    filter(Measurement.date > temp_start_date).\
    filter(Measurement.date < temp_end_date).all()

temps

average = temps[0][0]
maximum = temps[0][1]
minimum = temps[0][2]
```

```python
# plot past year temperature query

temps_df = pd.DataFrame(temps, columns = ['average', 'max', 'min'] )
temps_df.plot.bar(y=['average'], yerr=maximum-minimum)
plt.show()
```
